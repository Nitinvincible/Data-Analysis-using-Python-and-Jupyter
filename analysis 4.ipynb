'''1. Consider a list of values: 
bag = [25,26,21,22,31,29,33,34,26,30,31,46]
• Import the appropriate Python libraries to create a ndarray called bag_weights
having 3 rows and 4 columns from the list bag.<br>
• Use Numpy library to display the mean, variance and median of the given data in 
bag_weights.<br>
• Write a command to display the count of values greater than the median in 
bag_weights.<br>
• Transpose bag_weights and then split it in two arrays bagA and bagB having 2 
rows and three columns each.<br>
• Sort bagA such that it brings the highest value of the row in the first column. Sort bagB
such that it brings the lowest value of the row in the first column. <br>
• Find the union and intersection of values in bagA and bagB.<br>
'''

```python
import pandas as pd
import numpy as np
bag=[25,26,21,22,31,29,33,34,26,30,31,46]
bag_weights=np.array(bag)
bag_weights=np.reshape(bag_weights,(3,4))
print("#A. Bag_weight array:\n",bag_weights,'\n\n',
      "#B. Mean of bag_weight:\n",np.mean(bag_weights,axis=1),'\n\n',
           "Variance of bags_weights:\n",np.var(bag_weights),'\n\n',
           "Median of bags_weight:\n",np.median(bag_weights),'\n\n',
      "#C. Count of the values greater than the median:\n",(np.count_nonzero(bag_weights>np.median(bag_weights))),'\n')
print("#D. Transpose of bag_weight:\n ",np.transpose(bag_weights),'\n\n',)
[bagA,bagB]=np.split((np.transpose(bag_weights)),2)
print("bagA:\n",bagA,'\n\n',
      "bagB:\n",bagB,'\n')
#sorting bagA and bagB
#convertinng bagA to a 1 dimensional array then converting into a list 
bag1d=bagA.flatten()
lst=bag1d.tolist()
#using for loop and swaping to sort the list in descending order
n=0
for n in range (0,len(lst)):
    for p in range(n+1,len(lst)):
        if lst[n] <= lst[p]:
            lst[n],lst[p]=lst[p],lst[n]
print("Sorted bagA such that highest value in the first column:\n",np.array(lst).reshape(2,3),'\n\n',
      "sorted bagB:\n",np.sort(bagB))
#union and intersection of bagA and bagB
print("union of bagA and bagB:\n",np.union1d(bagA, bagB),'\n\n',
      "Intersection of bagA & bagB: \n", np.intersect1d(bagA,bagB))
```

'''2. Consider a list of values:<br>
rate = [4.23,3.8,2.98,2.56,3,114,3.8,3.78,2.98,4.8,4.10,3.65]<br>
• Import the appropriate Python libraries to create a one-dimensional ndarray called 
growth_rate from the list rate. Create another one-dimensional array named 
twos having the same number of elements as growth_rate, all set to 2.<br>
• Use Numpy library to find the index of the maximum and the minimum values in the 
array growth_rate.<br>
• What does a box plot show? Give a command to display a boxplot for growth_rate.<br>
• Concatenate the two arrays growth_rate and twos, and reshape the resulting 
array to have four rows and appropriate number of columns, call it results. <br>
• Find the mean, median, mode and standard deviation of each column in results.<br>
• Write a command to store the array results to a file called result.npy on the 
disk in the current working directory.'''

```python
import numpy as np
from scipy import stats as st
from matplotlib import pyplot as plt
rate = [4.23,3.8,2.98,2.56,3,114,3.8,3.78,2.98,4.8,4.10,3.65]
growth_rate=np.array(rate)
print("growth_rate:\n",growth_rate)
twos=np.full_like(np.ones([12]),[2])
print("#A. Another one dimensional array named twos: \n",twos,'\n')
print("#B. Indices of maximum element:\n",
        np.where(growth_rate == np.amax(growth_rate)),'\n\n',
      "Max element is:\n",np.amax(growth_rate),'\n\n',
      "Indices of minimum element:\n",
        np.where(growth_rate == np.amin(growth_rate)),'\n\n',
      "Minimum element is:\n",np.amin(growth_rate))
fig=plt.figure(figsize=(10,5))
plt.boxplot(growth_rate)
plt.show()
result=np.reshape(np.concatenate([growth_rate,twos]),(4,6))
print("#D. The concatenated Result:\n",result)
print("Mean:\n",np.mean(result, axis=0),'\n',
      "Median:\n",np.median(result,axis=0),'\n',
      "Mode:\n",st.mode(result,axis=0),'\n',
      "Standard deviation: \n",np.std(result,axis=0))
np.save("result.npy", result)
data=np.load('result.npy')
print("#E. The data inside the file:\n",data)
```

```python

import pandas as pd 
import numpy  as np
df={ 'Movie_title':['AAA', 'BBB','CCC','DDD','EEE','FFF','GGG','HHH','JJJ'],
        'Director_name':['Ram','Eash','Anuj','Jay','Eash','Suraj','Anuj','Ram','Anuj'],
        'Language':['Urdu','Hindi','Hindi','Hindi','Hindi','French','French','French','French'], 
        'Length':[120, np.nan, 125, 150, 90, 100, np.nan, 115, 120],
        'Budget':[90, 65, 100, 85, 60, 115, 80, 50, 92], 
        'Gross_collections':[80, 70,150 ,85, np.nan, 120, 81, 40, 75],
        'User_rating':[4, 6, 9, 6, 7, 8, 5, 3, 3],
        'Critic_rating':[7, 6, 8, 5, 5, 6, 5, 4, 6]}
print(pd.DataFrame(df))
new_df=(pd.DataFrame(df))
print(new_df)
row, column=new_df.shape
print("Number of  Rows:\n",row,'\n',"Number of Columns:\n",column,'\n')
having_null=(new_df.columns [new_df.isnull().any ()])
print("#B. Name of the columns having NULL: \n",having_null,'\n\n'
        "Count of the NULL value corresponding columns: \n",new_df.isna().sum(),'\n\n'
        "Replacing the NULL values from the lowest value of that column:\n",
        new_df.fillna(np.min(new_df)))
mean_of_col=(new_df['User_rating'].mean()+new_df['Critic_rating'].mean())/2
diff_col=new_df['Gross_collections']-new_df['Budget']
#including new columns
new_df=new_df.assign(Rating=mean_of_col,Profit=diff_col)
display(new_df)
#corelation between 2 columns
print(new_df['Budget'].corr(new_df['Rating'],method='pearson'),'\n',
"As this is showing nan that means bot columns are  independent, i.e non of them will varies if one of the varies\n\n")
grouped=new_df.groupby(['Director_name']).sum().head(1).sort_index()
display("Most profitable Director",grouped)
#crosstable
table=pd.crosstab(index=new_df['Director_name'],columns=new_df['Language'])
display(table)
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as  plt
iris=pd.read_csv("iris_dataset.csv")
display(iris)
print("#2.\nSummary of the  vital statistics with NAN;\n",iris.describe(include='all'),'\n\n',
        "Non numerical data: \n",iris.describe(include=object))
#dataframe without null rows
del_iris=iris.dropna()
#filling null values with mean 
rep_mean=iris.fillna(iris.mean())
#replacing null of  the column with the mode of the column
rep_mode=iris["petal length (cm)"].fillna(iris["petal length (cm)"].mode()[0])
display(del_iris)
display(rep_mean)
display(rep_mode)
iris_dummmy=pd.get_dummies(iris,['species'])
display(iris_dummmy)
display(iris.drop(columns=['species']))
#scatter plot
plt.scatter(iris['sepal_length_(cm)'],iris['petal length (cm)'],c='red',edgecolors='black',s=100)
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.title('scatter plot')
plt.rcParams.update({'figure.figsize':(8,6), 'figure.dpi':100})
plt.show()
#Histograms
plt.figure(figsize=(12,6))
plt.hist([iris['sepal_length_(cm)'],iris['sepal_width_(cm)']]
                ,bins=10,color=['red','blue'],edgecolor='black',alpha=0.5,label=['seplength','sepwidth'])
plt.legend(borderpad=0.5,fontsize=10)
plt.show()
plt.hist(iris['sepal_width_(cm)'],bins=5,color='blue',edgecolor='black',alpha=0.3,label='sepwidth')
plt.show()
#saving figures
plt.savefig('sep histo.png')
```
