```python
'''Q1: Create a series with all the names of colors in Rainbow:
a. Change the index values as multiples of 3.
b. Display the names of colors with an index value of more than 12.
c. Change the index values to “Violet”, “indigo”, ……, ”Black”, and “White”.
d. Check if there is any Null value in the series now.
e. Assign the index name colors to the NaN values occurred in section c.
'''
#created a series with all the names of colors in rainbow
import pandas as pd 
Rainbow=pd.Series(["Violet", "Indigo", "Blue", "Green", "Yellow", "Orange", "Red"], index=[0, 1, 2, 3, 4, 5, 6])
print(Rainbow,'\n')
#a using rename function to rename index
Rainbow_into3=Rainbow.rename(lambda x: x*3)
print(Rainbow_into3,'\n')
#b displaying names of the  colos whose  index values are greater tha 12 
gt12=Rainbow_into3[5:]
print(gt12,'\n')
#c
new_index=["Violet", "Indigo", "Blue", "Green", "Yellow", "Orange", "Red","Black",'White']
Rainbow.index=["Violet", "Indigo", "Blue", "Green", "Yellow", "Orange", "Red"]
Rainbow=pd.Series(Rainbow,index=new_index)
print(Rainbow,'\n')
#d checking null values
print(Rainbow.isnull(),'\n')
#e Assigning the nan as name of  index i.e colors
Rainbow["Black","White"]="Black","White"
print(Rainbow)
```

```python
'''Q3: Create a data frame of 10 students in your class having their Names, IDs, and Contact
Numbers.
a. Add a new column University ID to the data frame.
b. Delete the ID from the data frame.
c. Give the row names as one, two three…..
d. Find the values in the third row of the column 'Name'.
e. Print all the Contact details of students.
f. Add another column semester marks.
g. Add another column 'Result' where write 'pass' if semester marks are above 40 out of
100 else write 'fail'.'''
import pandas as pd
students=pd.DataFrame({"Name":['N1','N2','N3','N4','N5','N6','N7','N8','N9','N10'],
            "Ids":['ad1','ad2','ad3','ad4','ad5','ad6','ad7','ad8','ad9','ad10'],
            "Contact":[9899554112, 9625562642, 9955953694, 7838706126, 9065454408, 9953831170, 9205132112, 9899714103
                        ,9555556449, 8168131752]})
print(students,'\n')
# a adding university id as a column
'''df.insert(2, "university_id", [21, 23, 24, 21], True)'''
university_id=["x001","x002", "x003", "x004", "x005", "x006", "x007", "x008", "x009", "x010"]
students['University_id']=university_id
print(students,'\n\n')
#b Deleting IDs of  students
print(students.drop('Ids',axis=1),'\n')
#c assigning the rows names or  index 
df=students.set_axis(['One','Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten'],axis=0)
print(df.index,'\n',df,'\n')
#d extracting 
print(df.loc['Three',['Name']],'\n')
#e printing only  contact details
print(students[['Contact']].to_string(index=False),'\n')
#f adding new column semester marks
students.insert(4,"Semester marks",[98, 45, 23, 65, 11, 19, 24, 88, 69, 36], True)
print(students,'\n')
#g adding a conditionalcolumn where the results of students are shown
students['Results'] = ['Pass' if x >=40  else 'Fail' for x in students['Semester marks']]
print(students)
```

```python
'''Q4: Consider the following data dictionaries.
1. {‘a’: 12, ‘b’: 45, ‘d’: 79, ‘b’: 13, ‘a’: 43, ‘e’: 33}
2. {‘b’: 23, ‘d’:90, ‘f’: 55, ‘a’: 21, ‘s’: 26, ‘d’: 78}
Now perform the following operations:
a. Make series out of those two data dictionaries.
b. Append the two series by index.
c. Find the Union of series one and series two.
d. Find the difference between series one from two.
e. Compute the set intersection of the two series.
f. Delete index 4 from both series.'''
from typing import Set
import pandas as pd
import numpy as np
D1={'a': 12, 'b': 45, 'd': 79, 'b1': 13, 'a1': 43, 'e': 33}
D2={'b': 23, 'd':90, 'f': 55, 'a': 21, 's': 26, 'd1': 78}
#a. making Series out of those 2 data dict..
S1=pd.Series(D1)
S2=pd.Series(D2)
print(S1,'\n\n',S2,'\n')
#b Appending the two series by index 
S3=S1.append(S2).drop_duplicates()
print(S3,'\n')
#c Union of both
union=S1.combine_first(S2)
print (union,'\n')
#d Difference
print(S2-S1)
# creating an array  of different  elements from both series
print("The set difference is: ", np.setdiff1d(S2, S1))
# e Intersection
intersecion=pd.Series(list(set(S1).intersection(set(S2))))
print(intersecion)
#f  deleting elements from both having at index 3 i mean the the 4th number
print(S1.index,'\n',S2.index )
print("Elemnts of S1: ",'\n',S1.drop(index="b1"),'\n',"Elemnts of S2: ",'\n', S2.drop(index='a'))
```
